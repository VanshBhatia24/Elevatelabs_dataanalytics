```python
import numpy as np
import pandas as ps
import matplotlib.pyplot as plt #importing the chart and graph part of matplotlib
%matplotlib inline  #to show graphs in same page below the code line
import seaborn as sn
```

```python
df = ps.read_csv(r'C:\Users\vansh\Downloads\Python_Diwali_Sales_Analysis-main\Python_Diwali_Sales_Analysis-main\Diwali Sales Data.csv', encoding = 'unicode_escape') 
#To avoid any type of encoding error while reading the data, encoding is used
#We have to mention the full path of the csv file, if the location of jupyter python code file is not located in the same folder
#We have used r in front of the file location to resolve any error which might come due to special charactyers in file location

```

```python
#To drop columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
#axis=1 means full column, inplace=True means apply to the table
```

```python
df.dropna(inplace=True)
#Drops null value
```

```python
df.shape
```

```python
#Data type change
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
#returns arithmatic answers for columns with numberical data
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
# Gender Distribution of Buyers
sx=sn.countplot(x='Gender',hue='Gender',data=df,palette='pastel',legend=False)
for bars in sx.containers:
    sx.bar_label(bars)
```

```python
sales_power = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sn.barplot(x='Gender',hue='Gender',y='Amount',data=sales_power,palette='Set2', legend=False)
```

```python
# Gender Distribution Across Different Age Groups
ax=sn.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    #hue is used for categorisation of data furthermore
    #countplot is nothing but barplot with numbers
```

```python
#Total Amount vs Age Group Chart
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sn.barplot(x='Age Group',y='Amount',hue='Age Group',palette='Set2',data=sales_age,legend=False)
```

```python
sales_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sn.set(rc={'figure.figsize':(16,5)})
sn.barplot(x='State',hue='State', data=sales_state, y='Orders',palette='Set1',legend=False)
```

```python
ax=sn.countplot(data=df,x='Marital_Status',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sn.set(rc={'figure.figsize':(20,6)})

order = df['Occupation'].value_counts().index
ax=sn.countplot(x='Occupation',data=df, hue='Gender',order=order,palette='Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
order= df['Product_Category'].value_counts().index
ax= sn.countplot(x='Product_Category', data=df, order=order, hue='Product_Category', palette='Set1')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sn.pairplot(df[['Amount','Orders']])
plt.suptitle("Pairplot of Orders and Amount", y=1.03)
plt.show()
```

```python
ax= sn.histplot(df['Amount'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Purchase Amounts")
plt.ylabel("Frequency")
plt.show()
```
